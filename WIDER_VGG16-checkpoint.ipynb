```python
# Przerzuca wszystkie pliki ze zbiorów folderów do jednego folderu, usuwając poprzednie foldery.

def move_contents(input_folder):
    # Check if the input folder exists
    if not os.path.exists(input_folder):
        print(f"Folder '{input_folder}' does not exist.")
        return
    
    # Get the list of subfolders
    subfolders = [f for f in os.listdir(input_folder) if os.path.isdir(os.path.join(input_folder, f))]
    subfolders
    # Move the contents of each subfolder to the main folder
    for subfolder in subfolders:
        subfolder_path = os.path.join(input_folder, subfolder)
        # Check if the path is a folder
        if os.path.isdir(subfolder_path):
            # Get the contents of the subfolder
            subfolder_contents = os.listdir(subfolder_path)
            
            # Move the contents to the main folder
            for file in subfolder_contents:
                file_path = os.path.join(subfolder_path, file)
                destination_path = os.path.join(input_folder, file)
                shutil.move(file_path, destination_path)
            
            # Remove the empty subfolder
            os.rmdir(subfolder_path)
    
    print("Operation completed successfully.")

```

```python
def load_image(image_path):
    """
    Load an image from the specified path and return it as a pixel array.

    Parameters:
    - image_path (str): The path to the image file.

    Returns:
    - image (image): image.
    """
    try:
        # Open the image file
        image = Image.open(image_path)
        
        return image

    except Exception as e:
        print(f"Error loading image from {image_path}: {e}")
        return None

def read_description_file(file_path, images_to_process):
    # Function to read the description file and extract data for each image
    image_descriptions = []
    image_list = []
    
    with open(file_path, 'r') as file:
        while True:
            
            images_to_process -= 1
            
            if(images_to_process == -1):
                break
            
            # Read image path
            image_path = file.readline().strip()

            # Check if the file ends
            if not image_path:
                break

            # Read the value of k
            k = int(file.readline().strip())
            
            # If there are no faces then still in txt file is row with zeros
            if(k == 0):
                k = k + 1
            
            # Read k rows of 10 integers each, ignoring newline characters
            image_data = []
            for _ in range(k):
                row = file.readline().strip()
                if row:
                    image_data.append([int(x) for x in row.split()])

            # Append image description to the list
            image_descriptions.append({'image_path': image_path,'rows' : k, 'data': image_data})
            
            # Append image to image_list
            image_list.append(load_image(image_path))
            

    return image_descriptions, image_list
```

```python
def change_directory(folder_path):
    """
    Change the current working directory to the specified folder.

    Parameters:
    - folder_path (str): The path to the folder.

    Returns:
    - current_directory (str): The path to the current working directory after the change.
    """
    try:
        # Save the current working directory
        previous_directory = os.getcwd()

        # Change the current working directory
        os.chdir(folder_path)
        current_directory = os.getcwd()

        return current_directory, previous_directory

    except Exception as e:
        print(f"Error changing directory to {folder_path}: {e}")
        return None, None

def return_to_previous_directory(previous_directory):
    """
    Return to the previous working directory.

    Parameters:
    - previous_directory (str): The path to the previous working directory.

    Returns:
    - current_directory (str): The path to the current working directory after returning.
    """
    try:
        # Change back to the previous working directory
        os.chdir(previous_directory)
        current_directory = os.getcwd()
        return current_directory

    except Exception as e:
        print(f"Error returning to the previous directory: {e}")
        return None
```

# 2. Skalowanie zdjęć i augmentacja

## Skaluje zdjęcia do pownej wielkości, równej dla każdego zdjęcia.
## Dla każdego zdjęcia robię augmentację.

```python
def scale_images_and_descriptions(image_list, image_description, width, height):
    """
    Scale a list of PIL Images to the specified width and height.

    Parameters:
    - image_list (list): List of PIL Images.
    - width (int): Target width for scaling.
    - height (int): Target height for scaling.

    Returns:
    - scaled_images (list): List of scaled PIL Images.
    """
    scaled_images = []
    scaled_descriptions = []
    i = 0
    
    for image in image_list:
        try:
            
            description = image_description[i]
                
            # Scale the image
            scaled_image = image.resize((width, height))

            # Append the scaled image to the list
            scaled_images.append(scaled_image)
            
            rows = description['rows']
            
            j = 0 
            img_width, img_height = image.size

            while j < rows:
                row = description['data'][j]
                row[0] = row[0] / img_width * 100
                row[2] = row[2] / img_width * 100
                row[1] = row[1] / img_height * 100
                row[3] = row[3] / img_height * 100
                description['data'][j] = row[:4]
                j = j + 1
            
            scaled_descriptions.append(description)
            i = i + 1
        except Exception as e:
            print(f"Error scaling image: {e}")

    return scaled_images, scaled_descriptions
```

```python
# Tworze listę prostokatow dla twarzy
def get_labels(descriptions):
  
    size = len(descriptions)
    i = 0
    rectangle_cord = []
    rectangles = []
    
    while i < size:
        rectangle_cord.append(descriptions[i]['data'])
        rectangles.append(descriptions[i]['rows'])
        i = i + 1
        
    return rectangles, rectangle_cord

def images_to_array(images):

    # Zwróć listę obrazów w formie tablicy NumPy
    return np.array([np.array(image) / 255.0 for image in images])
```

## Augmentacja

```python
from PIL import Image, ImageEnhance, ImageOps
import numpy as np
import random

def augment_image_random(image_path, output_folder, num_images=10):
    # Otwórz oryginalne zdjęcie
    original_img = Image.open(image_path)

    # Utwórz folder wyjściowy, jeśli nie istnieje
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Generuj 10 losowo przekształconych wersji zdjęcia
    for i in range(num_images):
        # Klonuj oryginalne zdjęcie
        img = original_img.copy()

        # Losowe przekształcenia
        img = apply_random_transformations(img)

        # Zapisz przekształcone zdjęcie
        output_path = os.path.join(output_folder, f"augmented_{i+1}.jpg")
        img.save(output_path)

def apply_random_transformations(img):
    # Przekształcenie poziome (odbicie lustrzane)
    if random.random() > 0.5:
        img = ImageOps.mirror(img)

    # Przekształcenie pionowe (odbicie pionowe)
    if random.random() > 0.5:
        img = ImageOps.flip(img)

    # Zmiana jasności
    enhancer = ImageEnhance.Brightness(img)
    img = enhancer.enhance(random.uniform(0.5, 1.5))

    # Zmiana kontrastu
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(random.uniform(0.5, 1.5))

    # Zmiana nasycenia
    enhancer = ImageEnhance.Color(img)
    img = enhancer.enhance(random.uniform(0.5, 1.5))

    # Rotacja
    angle = random.uniform(-15, 15)
    img = img.rotate(angle, resample=Image.BICUBIC, expand=True)

    return img
```

# 4. Ćwiczenie Modelu

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, GlobalMaxPooling2D, Reshape, Concatenate
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models
import numpy as np
import tensorflow as tf
```

```python
def build_model(): 
    input_layer = Input(shape=(img_width, img_height, 3))
    
    vgg = VGG16(include_top=False)(input_layer)

    # Shared Features
    shared_features = GlobalMaxPooling2D()(vgg)
    
      # Classification Model  
    f1 = GlobalMaxPooling2D()(vgg)
    class1 = Dense(2048, activation='relu')(f1)
    class2 = Dense(1000, activation='softmax')(class1)

    # Bounding box model
    f2 = GlobalMaxPooling2D()(vgg)
    regress1 = Dense(2048, activation='relu')(f2)
    regress2 = Dense(1000 * 4, activation='sigmoid')(regress1)  

    # Reshape warstwy regress2 do (1000, 4)
    regress2_reshaped = Reshape((1000, 4))(regress2)

    facedetector = Model(inputs = input_layer, outputs = [class2, regress2_reshaped])
    
    return facedetector
```

```python
from tensorflow.keras.losses import categorical_crossentropy, mean_squared_error

def regression_loss(y_true, y_pred):
    
    regression_loss = 0
    
    # Dla regresji
    i=0
    while i < 4:
        regress_true = y_true[i]
        regress_pred = y_pred[i]
    
        regression_loss = regression_loss + mean_squared_error(regress_true, regress_pred)
        i = i + 1;

    return regression_loss
```

```python
class FaceTracker(Model): 
    def __init__(self, eyetracker,  **kwargs): 
        super().__init__(**kwargs)
        self.model = eyetracker

    def compile(self, opt, classloss, localizationloss, **kwargs):
        super().compile(**kwargs)
        self.closs = classloss
        self.lloss = localizationloss
        self.opt = opt
    
    def train_step(self, batch, **kwargs): 
        
        X, y = batch
        
        with tf.GradientTape() as tape: 
            classes, coords = self.model(X, training=True)
            
            batch_classloss = self.closs(y[0], classes)
            batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
            
            total_loss = batch_localizationloss + 0.5 * batch_classloss
            
            grad = tape.gradient(total_loss, self.model.trainable_variables)
        
        opt.apply_gradients(zip(grad, self.model.trainable_variables))
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
    
    def test_step(self, batch, **kwargs): 
        X, y = batch
        
        classes, coords = self.model(X, training=False)
        
        batch_classloss = self.closs(y[0], classes)
        batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
        total_loss = batch_localizationloss + 0.5 * batch_classloss
        
        return {"total_loss":total_loss, "class_loss":batch_classloss, "regress_loss":batch_localizationloss}
        
    def call(self, X, **kwargs): 
        return self.model(X, **kwargs)
```

```python
model = build_model()

classloss = tf.keras.losses.BinaryCrossentropy()
regressloss = regression_loss

batches_per_epoch = 32
lr_decay = (1./0.75 -1) / batches_per_epoch
opt = tf.keras.optimizers.Adam(learning_rate=0.001, decay=lr_decay)

model = FaceTracker(model)
model.compile(opt, classloss, regressloss)

# Trenowanie modelu
model.fit(train_images, (train_class_array, train_labels_array), epochs=10, batch_size = batches_per_epoch, validation_split=0.2)
```

```python
from tensorflow.keras.models import load_model, save_model

def save_custom_model(model, model_path):
    """
    Zapisuje wytrenowany model do pliku HDF5.

    Parameters:
    - model: Obiekt modelu do zapisu.
    - model_path: Ścieżka do pliku, do którego zostanie zapisany model.
    """
    save_model(model, model_path)
    print(f'Model został zapisany do: {model_path}')

def load_custom_model(model_path):
    """
    Wczytuje wytrenowany model z pliku HDF5.

    Parameters:
    - model_path: Ścieżka do pliku, z którego zostanie wczytany model.

    Returns:
    - Wytrenowany model wczytany z pliku.
    """
    model = load_model(model_path)
    print(f'Model został wczytany z: {model_path}')
    return model
```
